## **KMeans Clustering**

```python
df = pd.read_csv('crime_data.csv')
df
```

```python
df1=df.rename({"Unnamed: 0" : "State"},axis=1)
df1
```

```python
def norm_fun(i):
  x = (i-i.min())/(i.max()-i.min())
  return (x)
```

```python
surya2 = norm_fun(df1.iloc[:,1:])
surya2
```

```python
from sklearn.cluster import KMeans
clf = KMeans()
WCSS=[]

for i in range(1,11):
  clf = KMeans(n_clusters= i)
  clf.fit(surya2)
  WCSS.append(clf.inertia_)
plt.plot(range(1,11),WCSS)
plt.title('Elbow Curve')
plt.xlabel('no of clusters')
plt.ylabel('WCSS')
plt.show()


```

```python
clf = KMeans(n_clusters=4)
y_means = clf.fit_predict(surya2)
y_means
```

```python
clf.labels_
clf.cluster_centers_
```

```python
clf.inertia_
```

```python
Kmeans = pd.Series(y_means)
df1['Kmeans_clust'] = Kmeans
df1
```

```python
df1.groupby(df1.Kmeans_clust).mean()
```

```python
df1.plot(x='Murder',y = 'Assault',kind = "scatter",c = clf.labels_,s=50,cmap=plt.cm.coolwarm)
plt.title('Clusters Using KMeans')
```

```python
df1.plot(x='UrbanPop',y = 'Rape',kind = "scatter",c = clf.labels_,s=50,cmap=plt.cm.coolwarm)
plt.title('Clusters Using KMeans')
```

# **DB SCAN**

```python
vr = pd.read_csv('crime_data.csv')
vr
```

```python
vr1 = vr.rename({"Unnamed: 0" : "State"},axis =1)
vr1
```

```python
def norm_fun (i):
  x = (i-i.min())/(i.max()-i.min())
  return(x)
```

```python
df = norm_fun(vr.iloc[:,1:])
df
```

```python
v =df.values
v
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
#neigh = NearestNeighbors(n_neighbours=)
```

```python
from sklearn.cluster import DBSCAN
```

```python
dbscan= DBSCAN(eps=0.490,min_samples=5)
dbscan.fit(df)
```

```python
d = dbscan.labels_
d1=pd.Series(d)
```

```python
vr1['clust']=d1
vr1
```

```python
vr1.plot(x="UrbanPop",y ="Murder",c=dbscan.labels_ ,kind="scatter",s=50 ,cmap=plt.cm.copper_r)
plt.title('Clusters using DBScan')
```
