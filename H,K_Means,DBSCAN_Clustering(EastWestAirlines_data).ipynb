```python
from scipy.cluster.hierarchy import linkage
import scipy.cluster.hierarchy as sch
```

```python
z= (linkage(surya2,method = 'average',metric='euclidean'))
plt.figure(figsize=(15,6))
plt.title('hierarchy')
plt.xlabel('index')
plt.ylabel('distance')
sch.dendrogram(z)
plt.show()
```

```python
z= (linkage(surya2,method = 'complete',metric='euclidean'))
plt.figure(figsize=(15,6))
plt.title('hierarchy')
plt.xlabel('index')
plt.ylabel('distance')
sch.dendrogram(z)
plt.show()
```

## **KMeans Clustering**

```python
df = pd.read_csv('EastWestAirlines.csv')
df
```

```python
df1=df.rename({"Award?" : "Award"},axis=1)
df1
```

```python
df2 = df1.drop(['ID#'],axis=1)
df2
```

```python
def norm_fun(i):
  x = (i-i.min())/(i.max()-i.min())
  return (x)
```

```python
surya2 = norm_fun(df2.iloc[:,0:])
surya2
```

```python
from sklearn.cluster import KMeans
clf = KMeans()
WCSS=[]

for i in range(1,11):
  clf = KMeans(n_clusters= i)
  clf.fit(surya2)
  WCSS.append(clf.inertia_)
plt.plot(range(1,11),WCSS)
plt.title('Elbow Curve')
plt.xlabel('no of clusters')
plt.ylabel('WCSS')
plt.show()


```

```python
clf = KMeans(n_clusters=5)
y_means = clf.fit_predict(surya2)
y_means
```

```python
clf.labels_
clf.cluster_centers_
```

```python
clf.inertia_
```

```python
Kmeans = pd.Series(y_means)
df2['Kmeans_clust'] = Kmeans
df2
```

```python
df2.groupby(df2.Kmeans_clust).mean()
```

```python
clust_0 = df2[(df2.Kmeans_clust==0)]
clust_0
```

```python
clust_1 = df2[(df2.Kmeans_clust==1)]
clust_1
```

```python
clust_2 = df2[(df2.Kmeans_clust==2)]
clust_2
```

```python
clust_3 = df2[(df2.Kmeans_clust==3)]
clust_3
```

```python
clust_4 = df2[(df2.Kmeans_clust==4)]
clust_4
```

```python
vr = pd.read_csv('EastWestAirlines.csv')
vr
```

```python
vr2 = vr1.drop(['ID#'],axis =1)
vr2
```

```python
def norm_fun (i):
  x = (i-i.min())/(i.max()-i.min())
  return(x)
```

```python
df = norm_fun(vr2.iloc[:,0:])
df
```

```python
v =df.values
v
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
dbscan= DBSCAN(eps=0.90,min_samples=12)
dbscan.fit(df)
```

```python
dbscan.labels_
```
